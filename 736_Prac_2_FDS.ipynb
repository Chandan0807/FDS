# Winemag DataSet

```python
# Bar Chart
wine_reviews = pd.read_csv('Datasets/winemag-data-130k-v2.csv')
wine_reviews.head()
```

```python
# Bar Chart
# Create a figure and axis
fig, ax = plt.subplots()
# count the occurence of each class
data = wine_reviews['points'].value_counts()
# get x and y data
points = data.index
frequency = data.values
# create bar chart
ax.bar(points, frequency)
# set title and labels
ax.set_title("Wine review scores")
ax.set_xlabel("Points")
ax.set_ylabel("Frequency")
```

```python
import numpy as np 
import matplotlib.pyplot as plt
objects = ('Python','C++','Java','Perl','Scala','Lisp')
x_pos = np.arange(len(objects))
performance = [10,8,6,9,2,5]
# Bar Chart
# X Axis positions as first parameter list, it can be floating points
# Y values as 2nd parameter list
# Alpha is tranmsparency,
# Align can be center or edge
# Color can be single value or a list of color codes, one for each
plt.bar(x_pos, performance, width = 0.5, align = 'center', alpha = 1, color = ('r','r','g','g','b','b'))
# To define labels for x asis values.
plt.xticks(x_pos, objects)
plt.ylabel('Usage')
plt.xlabel('Programming languages')
plt.title('programming languages')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12,7))
countries = ['USA', 'Russia', 'Spain', 'India', 'UK', 'China']
totalDeaths = [33645,68445,55864,25565,47646,75415]
plt.bar(countries, totalDeaths, width=0.4, align='center',alpha=1, color= ('r','g','b','yellow','purple','orange','violet'))
i = 1.0
j = 2000

for i in range(len(countries)):
    plt.annotate(totalDeaths[i], (-0.1+i, totalDeaths[i]+j))

plt.legend(labels = ['Total Deaths'])
plt.title("Bar plot representing the total deaths by top 6 countries")
plt.xticks(countries, totalDeaths)
plt.ylabel('TotalDeaths')
plt.xlabel('Countries')
plt.savefig('1BarPlot.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,10))
countries = ['USA', 'Russia', 'Spain', 'India', 'UK', 'China']
totalCases = [335645,684445,565864,252565,478646,754315]
totalDeaths = [33645,68445,55864,25565,47646,75415]
for i in range(len(countries)):
    plt.bar(countries[i], totalDeaths[i], bottom= totalCases[i] - totalDeaths[i], color = 'black')
    plt.bar(countries[i], totalCases[i] - totalDeaths[i], color = 'r')
plt.legend(labels = ['Total Deaths', 'Total Cases'])
plt.ylabel('Total cases')
plt.xlabel('Countries')
plt.title("Bar plot representing the total deaths and total cases by top 12 countries")
plt.savefig('3BarPlot.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize=(15,10))
countries = ['USA', 'Russia', 'Spain', 'India', 'UK', 'China']
totalCases = [335645,684445,565864,252565,478646,754315]
totalDeaths = [33645,68445,55864,25565,47646,75415]
totalRecovery = [254654,545256,424453,167563,345614,641523]

X = np.arange(len(totalDeaths))

plt.bar(X, totalDeaths, color = 'black', width = 0.25)
plt.bar(X + 0.25, totalRecovery, color = 'g', width = 0.25)
plt.bar(X + 0.5, totalCases, color = 'b', width = 0.25)

plt.legend(['Total Deaths', 'Total Recovery', 'Total Cases'])

plt.xticks([i + 0.25 for i in range(6)], countries)

plt.title("Bar plot representing the total deaths, total recovery and total cases of top 6 countries")

plt.xlabel('Countries')
plt.ylabel('Cases')

plt.savefig('4BarPlot.png')

plt.show()
```

```python
from pandas import DataFrame
import matplotlib.pyplot as plt

Data = {'Tasks': [300,500,700],
       'Task Type': ['Tasks Pending', 'Tasks Ongoing', 'Tasks Completed']}

df = DataFrame(Data)
df.set_index('Task Type', inplace=True)

# autopct has extra % at the end as escape, as % is interpreted as formatting
# Only pie chart needs labels to be data frame index
df.plot.pie(y='Tasks', figsize=(10,10), autopct='%1.2f%%', startangle= 90)
```

```python
import numpy as np
import matplotlib.pyplot as plt
# if using Jupyter notebook include:
%matplotlib inline

# pue chart, where the slices will be ordered and plotted counter-clockwise:
labels = ['Civil', 'Electrical', 'Mechanical', 'Chemical']
sizes = [15,50,45,10]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.2f%%', startangle=90)
ax.axis('equal') # Equal aspects ratio ensures the pie chart is circular.
ax.set_title('Engineering Diciplines')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
# if using Jupyter notebook include:
%matplotlib inline

# pue chart, where the slices will be ordered and plotted counter-clockwise:
labels = ['Civil', 'Electrical', 'Mechanical', 'Chemical']
sizes = [15,50,45,10]

# Explode out the 'Chemcial' pie piece by offsetting it a greater amount
explode = (0.1,0.1,0.1,0.4)

fig, ax = plt.subplots()
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.2f%%', shadow=True, startangle=90)
ax.axis('equal') # Equal aspects ratio ensures the pie chart is circular.
ax.set_title('Engineering Diciplines')

plt.show()
```

# Subplot

```python
plt.figure(figsize=(20,10))
plt.subplot(2,2,1)
plt.bar(range(1,6), np.random.randint(1,20,5))
plt.title("2,2,1")

plt.subplot(2,2,2)
plt.bar(range(1,6), np.random.randint(1,20,5))
plt.title("2,2,2")

plt.subplot(2,2,3)
# s is the size of dot
plt.scatter(range(1,6), np.random.randint(1,20,5), s=200, color='r')
plt.title("2,2,3")

plt.subplot(2,2,4)
plt.plot(range(1,6), np.random.randint(1,20,5), marker='o', color='g', linestyle='--')
plt.title("2,2,4")
```
