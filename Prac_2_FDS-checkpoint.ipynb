```python
import pandas as pd 
df = pd.DataFrame({
    'names' : ['Chandan','Siddhesh','Bhagyawan','Gautham','Shivam'],
    'age' : [22,21,21,21,22],
    'gender' : ['M','M','M','M','M'],
    'state' : ['Maharashtra','Goa','Kerala','Chennai','Goa'],
    'num_children' : [0,2,0,1,2],
    'num_pets' : [1,2,3,1,2]
}) 
# From pandas to plots on same figure
# gca stands for 'get current axis'
ax=plt.gca()

df.plot(kind='line' ,x='names', y='num_children',ax=ax )
df.plot(kind='line' ,x='names', y='num_pets',ax=ax  )
```

```python
import pandas as pd 
iris = pd.read_csv('Datasets/Iris(1).csv')
print(iris.head())
```

```python
# get columns to plot
columns = iris.columns.drop(['Species', 'Id'])
# create x data
x_data = range(0, iris.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in columns:
    ax.plot(x_data, iris[column], label=column)
# set title and legend
ax.set_title('Datasets/Iris(1).csv')
ax.legend()
```

```python
# create figure and axis
fig, ax = plt.subplots()
# plot histogram
ax.hist(iris['SepalLengthCm'])
# set title and labels
ax.set_title("Iris")
ax.set_xlabel("Sepal_Length")
ax.set_ylabel("Frequency")
```

# Winemag DataSet

```python
# Bar Chart
wine_reviews = pd.read_csv('Datasets/winemag-data-130k-v2.csv')
wine_reviews.head()
```

```python
# Bar Chart
# Create a figure and axis
fig, ax = plt.subplots()
# count the occurence of each class
data = wine_reviews['points'].value_counts()
# get x and y data
points = data.index
frequency = data.values
# create bar chart
ax.bar(points, frequency)
# set title and labels
ax.set_title("Wine review scores")
ax.set_xlabel("Points")
ax.set_ylabel("Frequency")
```
